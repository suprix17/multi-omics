# Calculation of Sample specific Risk score

```python
rs_c = pd.DataFrame()
for i in df1.index.values:
    DF=DF1[i].astype(float)*df1.loc[i][1]
    rs_c[i] = DF.copy()
RS_C = rs_c.sum(axis=1)
RS_C = pd.DataFrame(RS_C, columns=['RS'])
C = pd.merge(DF1[['OS_MONTH' , 'OS_STATUS']], RS_C, left_index=True, right_index=True)
C.to_csv('RS_c.csv')
```
